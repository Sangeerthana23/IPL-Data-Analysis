## DATA CLEANING

## RENAMING COLUMNS

```python
a.rename(columns={"Playerss":"Player",
                 "Runss":"Runs",
                 "Strike Rate":"Strike_rate",
                 "Higest_score":"High_score",
                 "NO":"Not_Out"},inplace=True)
a
```

## REMOVING COLUMNS

```python
a.drop(columns="Unnamed: 0",inplace=True)
```

```python
a
```

## MODIFYING DATATYPES

```python
a['High_score'] = a['High_score'].str.replace('*', '', regex=False).astype(int)
a['Average'] = pd.to_numeric(a['Average'], errors='coerce')
    
```

```python
a.info()
```

## OUTLIERS

```python
plt.boxplot(a["Runs"])
```

## EXPLORATORY DATA ANALYSIS

## TOP RUN SCORERS

```python
top_scorers=a.groupby("Player")["Runs"].sum().sort_values(ascending=False).head(10)
```

```python
print(top_scorers)
```

## BAR CHART(TOP RUN SCORERS)

```python
plt.figure(figsize=(7,4))
sns.barplot(x=top_scorers.values,y=top_scorers.index, palette="magma")
plt.title("TOP SCORERS",fontsize=15)
plt.xlabel("Runs",fontsize=15)
plt.ylabel("Player",fontsize=15)
plt.tight_layout()
plt.show()
```

## TOTAL NUMBER OF SIXES BY TOP 10 PLAYERS

```python
sixes=a.groupby(by="Player")["Six"].sum().sort_values(ascending=False).head(10)
```

```python
print(sixes)
```

## PIE CHART(TOTAL NUMBER OF SIXES BY TOP 10 PLAYERS)

```python
plt.figure(figsize=(5,5))
sixes.plot(kind='pie',autopct='%1.1f%%', startangle=90,colors=plt.cm.Paired.colors, textprops={"fontsize":8,}
              ,wedgeprops={"linewidth":0.5,'edgecolor':'black'})
plt.title("TOP 10 SIX HITTERS â€“ % DISTRIBUTION",fontsize=15)
plt.ylabel('Six',fontsize=15)
plt.show()
```

## TOTAL NUMBER OF FOURS BY TOP 10 PLAYERS

```python
fours=a.groupby(by="Player")["Four"].sum().sort_values(ascending=False).head(10)
```

```python
print(fours)
```

## BARH CHART(TOTAL NUMBER OF FOURS BY TOP 10 PLAYERS)

```python
plt.figure(figsize=(7,4))
plt.barh(fours.index[::-1], fours.values[::-1],color='grey',edgecolor="black")
plt.title("TOP 10 PLAYERS BY FOURS (2008â€“2022)",fontsize=15,color="grey")
plt.xlabel("No Of Fours",fontsize=15,color="grey")
plt.ylabel("Players",fontsize=15,color="grey")
plt.tight_layout()
plt.show()
```

## TOTAL NUMBER OF NOTOUT BY TOP 10 PLAYERS

```python
not_out=a.groupby("Player")["Not_Out"].sum().sort_values(ascending=False).head(10)
```

```python
print(not_out)
```

## LINE CHART(TOTAL NUMBER OF NOTOUT BY TOP 5 PLAYERS)

```python
plt.figure(figsize=(7,4))
not_out.plot(kind="line",color='green')
plt.title("TOP 10 PLAYERS BY NOT OUT (2008â€“2022)",fontsize=15,color="green")
plt.xlabel("Players",fontsize=15,color="green")
plt.ylabel("No Of Notout",fontsize=15,color="green")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## TOP 10 STRIKE RATE OF A BATSMEN (MINIMUM 10 MATCHES PLAYED)

```python
strikerate=a[a["Matches"]<=10].groupby("Player")["Strike_rate"].mean().sort_values(ascending=False).head(10)
```

```python
print(strikerate)
```

## BAR PLOT(STRIKE RATE OF A BATSMEN)

```python
plt.figure(figsize=(7,4))
sns.barplot(strikerate,palette="viridis")
plt.title("TOP 10 PLAYERS BY STRIKE RATE (Min 10 Matches)",fontsize=15)
plt.xlabel("Players",fontsize=15)
plt.ylabel("Average Strike Rate",fontsize=15)
plt.tight_layout()
plt.show()
```

## MOST NUMBER OF FIFTY BY TOP 10 PLAYERS

```python
fifties=a.groupby("Player")["Fifty"].sum().sort_values(ascending=False).head(10)
```

```python
print(fifties)
```

## SCATTER PLOT(NUMBER OF FIFTIES BY TOP 10 PLAYERS)

```python
plt.figure(figsize=(7,4))
sns.scatterplot(x=fifties.values,y=fifties.index,color="purple")
plt.title("NUMBER OF FIFTIES BY TOP 10 PALYERS",fontsize=15,color='purple')
plt.xlabel("Count of Fifties",fontsize=15,color="purple")
plt.ylabel("Players",fontsize=15,color="purple")
ax=plt.gca()
ax.set_facecolor("silver")
plt.tight_layout()
plt.show()
```

## TOP 10 AVERAGE OF A BATSMEN (MINIMUM 20 MATCHES PLAYED)
